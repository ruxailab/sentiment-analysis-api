##### Tiny

```python
import torch
from transformers import pipeline

model_tiny = "openai/whisper-tiny"
device = 0 if torch.cuda.is_available() else "cpu"

pipe_tiny = pipeline(
    task="automatic-speech-recognition",
    model=model_tiny,
    chunk_length_s=30,
    device=device,
)

audio='/content/temp.mp3'

out_tiny = pipe_tiny(audio, return_timestamps=True)
# Save in json
import json
with open('/content/out_tiny.json', 'w') as f:
    json.dump(out_tiny, f)

print(out_tiny)
```
